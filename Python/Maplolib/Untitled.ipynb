```python
 from matplotlib import pyplot as plt
import pandas as pd

csvfile = pd.read_csv("countries.csv")

csvfile
```

```python
csvfile.year 
```

```python
china = csvfile[csvfile.country == "China"]
china
```

```python
plt.plot(china.year, china.population/10**6)
plt.show()
```

```python
gha = csvfile[csvfile.country == "Ghana"]
gha
```

```python
plt.plot(gha.year, gha.population/1000)
plt.show()
```

```python
plt.plot(china.year, china.population/10**6)
plt.plot(china.year, gha.population/10**5)

plt.title("Comparing the population of Ghana and China")
plt.xlabel("years")
plt.ylabel("Population growth")
plt.legend(["China","Ghana"])
plt.show()
```
